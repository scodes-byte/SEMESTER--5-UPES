#T-est Visualization:-
- **One-Sample t-test:** Compare a sample mean to a population mean.  
- **Independent Two-Sample t-test:** Compare means of two independent groups.  
- **Paired Sample t-test:** Compare means of two related samples (before vs after).  
- **Visualizations:** Histogram, Boxplot, and Line plot show data distribution and differences.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")  # nice plot style
%matplotlib inline
```

```python
# Dataset 1: Group A
group_a = np.random.normal(loc=75, scale=5, size=20)  # mean=75, std=5

# Dataset 2: Group B
group_b = np.random.normal(loc=80, scale=5, size=20)  # mean=80, std=5

print("Group A:", group_a)
print("Group B:", group_b)
```

```python
# Cell 3: Independent Two-Sample t-test
# Sample data
group_a = np.array([78, 82, 69, 74, 80])
group_b = np.array([70, 65, 72, 68, 75])

# Independent t-test
t_stat_2, p_val_2 = stats.ttest_ind(group_a, group_b)
print("Independent Two-Sample t-test:")
print(f"T-statistic: {t_stat_2:.2f}, P-value: {p_val_2:.4f}")

# Visualization
plt.figure(figsize=(6,4))
sns.boxplot(data=[group_a, group_b], palette='pastel')
plt.xticks([0,1], ['Group A', 'Group B'])
plt.title("Independent Two-Sample t-test Visualization")
plt.ylabel("Scores")
plt.show()
```

```python
# Cell 4: Paired Sample t-test
# Sample data: pre-test and post-test scores
pre_test = np.array([78, 82, 69, 74, 80])
post_test = np.array([80, 85, 72, 76, 83])

# Paired t-test
t_stat_3, p_val_3 = stats.ttest_rel(pre_test, post_test)
print("Paired Sample t-test:")
print(f"T-statistic: {t_stat_3:.2f}, P-value: {p_val_3:.4f}")

# Visualization
plt.figure(figsize=(6,4))
plt.plot(pre_test, marker='o', label='Pre-test', color='blue')
plt.plot(post_test, marker='o', label='Post-test', color='green')
for i in range(len(pre_test)):
    plt.plot([i, i], [pre_test[i], post_test[i]], color='gray', linestyle='--', alpha=0.6)
plt.xticks(range(len(pre_test)), [f'Student {i+1}' for i in range(len(pre_test))])
plt.title("Paired Sample t-test Visualization")
plt.ylabel("Scores")
plt.legend()
plt.show()
```
